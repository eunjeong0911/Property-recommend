```python
import pandas as pd
import os

# 1) 경로 설정
BASE_DIR = r"C:/dev/study/eunjeong/SKN18-FINAL-1TEAM"
DATA_DIR = os.path.join(BASE_DIR, "data", "raw")

INPUT_PATH  = os.path.join(DATA_DIR, "match_data.csv")
OUTPUT_PATH = os.path.join(DATA_DIR, "match_data_clean.csv")  # 덮어쓰고 싶으면 match_data.csv 로 바꿔도 됨

print("원본 파일 :", INPUT_PATH)
print("저장 경로 :", OUTPUT_PATH)

# 2) 데이터 로드
df = pd.read_csv(INPUT_PATH, encoding="utf-8-sig")
print("원본 shape:", df.shape)
```

```python
# 지우고 싶은 컬럼 리스트
drop_cols = [
    "등록번호_norm",
    "중개사명_norm",
    "대표자_norm",
    "사업자상호_norm",
    "중개업자명_norm",
    "상호_norm",
    "addr_key",
    "ldCode",
    "sttusSeCodeNm",
    "mnnmadr",
    "rdnmadr",
    "sttusSeCode",
    "bsnmCmpnm",
]


# 해당 컬럼 삭제
df = df.drop(columns=drop_cols)
```

```python
# 4) 실제로 존재하는 컬럼만 삭제 (없으면 에러 안 나게)
drop_exist = [c for c in drop_cols if c in df.columns]
print("삭제할 컬럼들:", drop_exist)

df = df.drop(columns=drop_exist)

print("삭제 후 shape:", df.shape)

# 5) 저장
df.to_csv(OUTPUT_PATH, index=False, encoding="utf-8-sig")
print("✅ 컬럼 삭제 후 저장 완료 →", OUTPUT_PATH)
```
